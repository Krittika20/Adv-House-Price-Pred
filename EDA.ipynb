```python
#histograms help us to see skeweness of our dataset, whether dataset is positively skewed, negatively skewed or 
#has normal distribution. It further help us to identify if there is need to remove outliers in our dataset.
housing = housing.drop(["Id"], axis = 1)
housing.hist(figsize =(20, 15))
plt.show()
```

From above histogram we can infer that most of the variables are positively skewed. 
Even the sale price that we need to predict is positively skewed too.

```python
import seaborn as sns
y = housing["SalePrice"]
df = housing.drop(["SalePrice"], axis = 1)

count = 1
plt.subplots(figsize=(15,60))
for i in df.columns:
    if(df[i].dtype == np.float64 or df[i].dtype == np.int64):
        plt.subplot(12,3, count)
        sns.scatterplot(x = df[i],y = y)
        count += 1
plt.show()
    
```

From above scatter plot we can say that most features are not much correlated to SalePrice like YrSold but features like OverallQuality and LotFrontage are highly correlated

```python
import seaborn as sns
y = housing["SalePrice"]
df = housing.drop(["SalePrice"], axis = 1)

count = 1
plt.subplots(figsize=(20,40))
for i in df.columns:
    if(df[i].dtype == np.object):
        plt.subplot(11,4, count)
        g = sns.barplot(x = df[i],y = y)
        #g.set_xticklabels(g.get_xticklabels(), rotation=20)
        count += 1
plt.show()
```

From above bargraph that is plot between categorical data and varying central tendancy of saleprice we can infer that many features such as CentralAir, HeatingHQ,KitchenQual ,NeighbourhooD etc have good effect on SP whereas features such as roofstyle has little effect on SP.

# Missing values

```python
ser = housing.isnull().sum()
df = pd.DataFrame({'Feature': ser.index, 'null_num': ser.values})
```

```python
plt.figure(figsize= (20,5))
sns.barplot(x = df.Feature, y = df.null_num)
plt.xticks(rotation = 90)
plt.show()
```

We can see there are some 6 features with many missing values we need to furthur deal with them.
